```python
import os 
import shutil
```

# move from subfolders to one big folder

```python
destination_folder = r"C:\Users\Saba\Desktop\Python\vigo\AIMV\data\images/"
images_start_folder = r'C:\Users\Saba\Desktop\Python\vigo\AIMV\jpgs/'
```

```python
for folder in os.listdir(images_start_folder):
    # print(folder)
    for subfolder in os.listdir(images_start_folder+folder):
        for filename in os.listdir(images_start_folder+folder+f"/{subfolder}"):
            if filename.lower().endswith(('.png', '.jpg', '.jpeg')):
                path_to_image = images_start_folder+f"{folder}/{subfolder}/{filename}"
                print(path_to_image)
                shutil.copy2(path_to_image, destination_folder)
                
```

### copy label names to /data/labels folder and leave only useful ones

```python
bounding_box_labels_folder_path = "./bounding_box_labels/"
labels_destination_folder = "./data/labels/"
set_of_image_names = set(os.listdir(destination_folder))
# print(set_of_image_names)
for file in os.listdir(bounding_box_labels_folder_path):
    
    if file[:-4]+".jpg" in set_of_image_names:
        shutil.copy2(bounding_box_labels_folder_path+file, labels_destination_folder+file)
    
```

## changing labelImg labeled values to 0

```python
folder = r'C:\Users\Saba\Desktop\51_geo_esn\labels/'
for file in os.listdir(folder):
    with open(folder+file,"r") as f:
        contents = f.read()
        # print(contents)
        contents_mod = contents[:4].replace("15",'0')+contents[4:]
    with open(f"{folder}/labels_modified/{file}", 'w') as f_mod:
        f_mod.write(contents_mod)
```
